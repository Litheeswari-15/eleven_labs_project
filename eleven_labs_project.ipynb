```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("d.csv")
```

```python
df.columns = df.columns.str.strip()

print("="*60)
print("FIRST 5 ROWS")
print(df.head())

print("\n" + "="*60)
print("DATASET INFORMATION")
print(df.info())

print("\n" + "="*60)
print("MISSING VALUES")
print(df.isnull().sum())

print("\n" + "="*60)
print("STATISTICAL SUMMARY")
print(df.describe())
```

```python
print("\n" + "="*60)

print("Total Pitches :", len(df))

print("Total Deals Made :", (df["deal"] == "Deal").sum())

print("No Deals :", (df["deal"] != "Deal").sum())

print("Total Investment Amount :",
      df["deal_amount"].sum())

print("Average Investment :",
      round(df["deal_amount"].mean(),2))

print("Highest Investment :",
      df["deal_amount"].max())

print("Lowest Investment :",
      df["deal_amount"].min())
```

```python
print("\nTOP 10 HIGHEST INVESTMENTS")

top = df.sort_values(
    by="deal_amount",
    ascending=False
)[["brand_name","deal_amount"]]

print(top.head(10))
```

```python
print("\nTOP BUSINESS IDEAS")

print(df["idea"].value_counts().head(10))
```

```python
print("\nTOTAL SHARKS INVESTED")

print(df["total_sharks_invested"].value_counts())
```

```python
print("\nAVERAGE DEAL AMOUNT")

print(df.groupby("deal")["deal_amount"].mean())
```

```python
plt.figure(figsize=(10,5))

top10 = df.sort_values(
    "deal_amount",
    ascending=False
).head(10)

plt.bar(
    top10["brand_name"],
    top10["deal_amount"]
)

plt.xticks(rotation=45)

plt.title("Top 10 Investments")

plt.xlabel("Brand")

plt.ylabel("Investment Amount")

plt.tight_layout()

plt.savefig("top10_investments.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

df["deal_amount"].plot(
    kind="hist",
    bins=20
)

plt.title("Investment Distribution")

plt.xlabel("Deal Amount")

plt.tight_layout()

plt.savefig("investment_distribution.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

df["deal"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")

plt.title("Deals vs No Deals")

plt.tight_layout()

plt.savefig("deal_pie_chart.png")

plt.show()
```

```python
plt.figure(figsize=(10,5))

df["idea"].value_counts().head(10).plot(
    kind="bar"
)

plt.title("Top Business Ideas")

plt.xlabel("Business Idea")

plt.ylabel("Count")

plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig("top_business_ideas.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

df["total_sharks_invested"].value_counts().sort_index().plot(
    kind="bar"
)

plt.title("Number of Sharks Invested")

plt.xlabel("Sharks")

plt.ylabel("Number of Deals")

plt.tight_layout()

plt.savefig("sharks_invested.png")

plt.show()

print("\nAnalysis Completed Successfully!")
```
